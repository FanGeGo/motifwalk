```python
polblogs_all = GraphContainer(find_meta("polblogs"), dloc)
```

```python
pol_labels = polblogs_all.get_labels()
label0 = np.where(pol_labels[:,0] == 1)[0]
label1 = np.where(pol_labels[:,0] == 0)[0]
pol_gt = polblogs_all.get_gt_graph()
l0_filter = pol_gt.new_vertex_property("bool")
l1_filter = pol_gt.new_vertex_property("bool")
for i in label0:
    l0_filter[i] = True
for i in label1:
    l1_filter[i] = True

l0_g = GraphView(pol_gt, l0_filter)
l1_g = GraphView(pol_gt, l1_filter)
graph_draw(l0_g)
print("Label 0 - Polblogs")
graph_draw(l1_g)
print("Label 1 - Polblogs")
```

```python
from motifwalk.motifs import all_3
# Label 0
for i, mc in enumerate(all_3):
    graph_draw(mc.gt_motif, output_size=(100,100))
    print("Motif index: {}".format(i))
    idx = l0_g.vertex_index.copy("int")
    shuffle(idx.a)
    g = Graph(l0_g, vorder=idx)
    print(ms(g, k=3, n_shuffles=10, motif_list=[mc.gt_motif])[1])
```

```python
# Label 1
for i, mc in enumerate(all_3):
    graph_draw(mc.gt_motif, output_size=(100,100))
    print("Motif index: {}".format(i))
    idx = l1_g.vertex_index.copy("int")
    shuffle(idx.a)
    g = Graph(l1_g, vorder=idx)
    print(ms(g, k=3, n_shuffles=10, motif_list=[mc.gt_motif])[1])
```

```python
inter_edge_filter = pol_gt.new_edge_property("bool")
for e in pol_gt.edges():
    if e.source() in label0:
        if e.target() in label1:
            inter_edge_filter[e] = True
    else:
        if e.target() in label0:
            inter_edge_filter[e] = True
```

```python
li_g = GraphView(pol_gt, efilt=inter_edge_filter)
graph_draw(li_g)
```
